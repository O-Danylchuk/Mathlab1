```python
# TASK 1
import numpy as np
import matplotlib.pyplot as plt 

shapeA = np.array([[0, 1], [1, 4], [3, 5], [1, 0], [0, 1]])
plt.plot(shapeA[:, 0], shapeA[:, 1]) # separated x & y coordinates for proper output

# naming
plt.title('Shape A')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.axis('equal')  # making x-axis to grow the same, as y-axis grows

plt.show()

shapeB = np.array([[-1, 0], [-1, 7.5], [2, 7.5], [0, 3.5], [0, 2], [-1, 0]])
plt.plot(shapeB[:, 0], shapeB[:, 1])
# naming
plt.title('Shape B')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.axis('equal') 

plt.show()
```
